```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load
#Machine Learning Programming 2

import random 
import numpy as np 
import pandas as pd
from sklearn.metrics import confusion_matrix 

counter_spam = 0.0
counter_not_spam = 0.0

data = pd.read_csv("../input/spammy/spambase.data",header=None, dtype=float)
data.head()
data = np.array(data)
m, n = data.shape


print('rows')
print(m)
print('----')
print('cols')
print(n)
print('----')
np.random.shuffle(data)

#split data
print('after splitting data')
train_data = data[:2300,:]
test_data = data[2300:,:]
print(train_data.shape)
print('----')
print(test_data.shape)
```

```python
#compute percentage of spam/not spam in our dataset
train_target = train_data[:,57]
test_target = test_data[:,57]
target_size = len(train_target)
print(train_target.shape)

for i in range(0,train_target.shape[0]):
    if(train_target[i]== 1):
        #print(flat_target[i])
        counter_spam += 1
    else:
        #print(flat_target[i])
        counter_not_spam += 1
prior_spam = counter_spam/target_size
prior_not_spam = counter_not_spam/target_size
print('Prior Probablities:')
print("Spam: ",prior_spam)
print('Not Spam: ',prior_not_spam)
```

```python
#probability of features given a class, etc.
#compute mean for each feature given a class, mean of feature2 for spam/notspam, class1/class2
spam_mean = []
not_spam_mean = []
spam_deviation = []
not_spam_deviation = []
print('number of COLS trainingSet')
print(train_data.shape[1])
rows = train_data.shape[0]
cols = train_data.shape[1]
s = []
ns = []
for row in range(rows):
    for col in range(cols):
        #store spam features
        if (train_data[row][57] == 1):
            #print(s)
            s.append(train_data[row][col])
        #store not_spam features
        else:
            ns.append(train_data[row][col])  
    spam_mean.append(np.mean(s))
    spam_deviation.append(np.std(s))
    not_spam_mean.append(np.mean(ns))
    not_spam_deviation.append(np.mean(ns))
    #reinitialize our spam/notSpam arrays because we want probablity averages of
    #features for each column, not for all combined columns
    #print(len(s))
    s = []
    ns = []

#print(len(spam_mean))
#print(len(not_spam_mean))
#print(len(spam_deviation))
#print(len(spam_deviation))
```

```python
#gaussian formula
def gaussNB(test_data,std, mean):
    #np.seterr(divide='ignore')
    term1 = float(1/(std*(np.sqrt(2*np.pi))))
    term2 = float(np.exp(-(np.square(test_data - mean))/(2 * np.square(float(std*std)))))
    return term1*term2
    
```

```python
#Classify/test model
result=[]
rows = test_data.shape[0]
cols = test_data.shape[1]
for i in range(rows):
    post_prob_spam = np.log(prior_spam)
    post_prob_not_spam = np.log(prior_not_spam)
    for j in range(cols):
        post_prob_spam += np.log(gaussNB(test_data[i][j], spam_deviation[j],spam_mean[j] ))
        post_prob_not_spam += np.log(gaussNB(test_data[i][j], not_spam_deviation[j],not_spam_mean[j] ))
    toClassify = [post_prob_spam,post_prob_not_spam]
    classify = np.argmax(toClassify)
    result.append(classify)
print(len(result))
```

```python

#confusion matrix
matrix = confusion_matrix(test_target,result)
print('confusion matrix:')
print(matrix)
true_positive = 0.0
true_negative = 0.0
false_positive = 0.0
false_negative = 0.0 
classification = 0

for i in range(2301):
    if (result[i] == 1 and test_target[i] == 1):
        true_positive+=1
    if (result[i] == 0 and test_target[i] == 0):
        true_negative +=1
    if (result[i] == 1 and test_target[i] == 0):
        false_positive +=1
    if (result[i] == 0 and test_target[i] == 1):
        false_negative +=1
        
        
#get accuracy, precision and recall

acc = float(true_positive + true_negative) / (true_positive + true_negative + false_negative + false_positive)
print("acc: ",acc)

prec = float(true_positive) / (true_positive + false_positive)
print("prec: ",prec)

recall = float(true_positive) / (true_positive + false_negative)
print("recall: ",recall)

    
```
